# video edge detection


```python
def Cartoon(image_color):
    output_img=cv2.stylization(image_color,sigma_s=40,sigma_r=0.3)
    return output_img
```

```python
def LiveCamEdgeDetection_canny(image_color):
    threshold_1=20
    threshold_2=50
    image_gray=cv2.cvtColor(image_color,cv2.COLOR_BGR2GRAY)
    canny=cv2.Canny(image_gray,threshold_1,threshold_2)
    return canny
```

```python
import time
cap=cv2.VideoCapture(0)
time.sleep(2)
while True:
    #capture frame by frame
    ret,frame=cap.read()
    print('ret',ret)
    
    if ret==False:
        break
        
    #display the resulting frame
    cv2.imshow('edge detection',LiveCamEdgeDetection_canny(frame))
    cv2.imshow('webcam',frame)
    
    
    #this command let us quit with 'q' button on a keyboard
    
    if cv2.waitKey(1) & 0xFF ==ord('q'):
        break
        
cap.release()

cv2.destroyAllWindows()            
```
