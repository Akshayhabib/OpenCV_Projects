# connecting to webcam

```python
import cv2
import numpy as np
import time
```

```python
cap=cv2.VideoCapture(0)

time.sleep(2)
width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
while True:
    #capture frame by frame
    ret,frame=cap.read()
    print('ret',ret)
    
    if ret==False:
        break
        
    #display the resulting frame
    cv2.imshow('frame',frame)
    
    #this command let us quit with 'q' button on a keyboard
    
    if cv2.waitKey(1) & 0xFF ==ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()

        
import time
```

# grayscale

```python
cap=cv2.VideoCapture(0)

time.sleep(2)
width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
while True:
    #capture frame by frame
    ret,frame=cap.read()
    print('ret',ret)
    
    if ret==False:
        break
    grayImg = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)    
    #display the resulting frame
    cv2.imshow('frame',frame)
    cv2.imshow('gray',grayImg)
    
    #this command let us quit with 'q' button on a keyboard
    
    if cv2.waitKey(1) & 0xFF ==ord('q'):
        break
        
cap.release()

cv2.destroyAllWindows()        
```

# video

```python
cap=cv2.VideoCapture('DATA-20220515T070511Z-001/DATA/vid2.mp4')
time.sleep(2)
width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
while True:
    #capture frame by frame
    ret,frame=cap.read()
    print('ret',ret)
    
    if ret==False:
        break
    img=cv2.resize(frame,(960,540))
    (B,G,R)=cv2.split(img)
    z=np.zeros(img.shape[:2],dtype='uint8')
    
    cv2.imshow('red',R)
    cv2.imshow('green',G)
    cv2.imshow('blue',B)
    
    cv2.imshow('red1',cv2.merge([z,z,R]))
    cv2.imshow('green1',cv2.merge([z,G,z]))
    cv2.imshow('blue',cv2.merge([B,z,z]))
   
    #display the resulting frame
    cv2.imshow('og',img)
    
    
    #this command let us quit with 'q' button on a keyboard
    
    if cv2.waitKey(1) & 0xFF ==ord('q'):
        break
        
cap.release()

cv2.destroyAllWindows()        
```

```python
cap=cv2.VideoCapture('DATA-20220515T070511Z-001/DATA/vid2.mp4')
time.sleep(2)
width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
frames=int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print('w-{} h-{} f-{}'.format(width,height,frames))
writer=cv2.VideoWriter('C:/Users/akshay0909/open CV/DAY2/DATA-20220515T070511Z-001/DATA/my_capture.mp4',cv2.VideoWriter_fourcc(*'XVID'),15,(width,height)) #15--fps
while True:
    #capture frame by frame
    ret,frame=cap.read()
    print('ret',ret)
    
    if ret==False:
        break
    img=cv2.resize(frame,(960,540))
    (B,G,R)=cv2.split(img)
    z=np.zeros(img.shape[:2],dtype='uint8')
    
    red_ch=cv2.merge([z,z,R])
    cv2.imshow('red',red_ch)
    
    #write the video
    writer.write(red_ch)

    #display the resulting frame
    cv2.imshow('frame',red_ch)
    
    
    #this command let us quit with 'q' button on a keyboard
    
    if cv2.waitKey(1) & 0xFF ==ord('q'):
        break
        
cap.release()

cv2.destroyAllWindows()        
```
