# Choose any video and process it by splitting the 3 channels and display all the 3 channels. 

```python
import cv2
import numpy as np
import glob
import matplotlib.pyplot as plt
import time
```

```python
cap=cv2.VideoCapture(r"C:\Users\akshay0909\ML & Dl projects\IMDB_NLP\Goku and Frieza Vs Jiren 60FPS.mp4")
time.sleep(2)
width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
while True:
    #capture frame by frame
    ret,frame=cap.read()
    print('ret',ret)
    
    if ret==False:
        break
    img=cv2.resize(frame,(360,240))
    (B,G,R)=cv2.split(img)
    z=np.zeros(img.shape[:2],dtype='uint8')
    
   
    cv2.imshow('red1',cv2.merge([z,z,R]))
    cv2.imshow('green1',cv2.merge([z,G,z]))
    cv2.imshow('blue',cv2.merge([B,z,z]))
   
    #display the resulting frame
    cv2.imshow('og',img)
    
    
    #this command let us quit with 'q' button on a keyboard
    
    if cv2.waitKey(5) & 0xFF ==ord('q'):
        break
        
cap.release()

cv2.destroyAllWindows() 
```
